```python
import tellurium as te
import matplotlib.pyplot as plt

model = te.loada("""
model repressilator
// Species 
mA = 0; //mrna A
mB = 0; //mrna B
mC = 0; //mrna C
A = 10; //protein A
B = 0; //protein B
C = 0; //protein C

                 
// Parameters
k1 = 50; // activated transcription
k2 = 0.001; // basal transcription
k3 = 5; // translation rate
n = 3; // hill coefficient
K = 40; // Repression threshold
k_deg_m = 0.2; // Degradation rate of mRNA
k_deg_p = 0.2; // Protein degradation
                 

// Reactions (transcription of mRNA)
J1: -> mA; k1/(1+(C/K)^n+k2)
J2: -> mB; k1/(1+(A/K)^n+k2)
J3: -> mC; k1/(1+(B/K)^n+k2)
                 
// Translation
J4: -> A; k3*mA
J5: -> B; k3*mB
J6: -> C; k3*mC
                 
// Degradation
J7: mA -> ; k_deg_m*mA
J8: mB -> ; k_deg_m*mB
J9: mC -> ; k_deg_m*mC
J10: A -> ; k_deg_p*A
J11: B -> ; k_deg_p*B
J12: C -> ; k_deg_p*C
end
""")

model.simulate(0, 500, 1000)
model.plot(title="Repressilator Oscillator Circuit")
plt.savefig("Repressilator Oscillator.png", dpi=300)
```
